### Stores match results for top 5 2025/2026 season NPFL teams in a dictionary and computes a standings table using only Python variables and arithmetic.
**Data Source: The NPFL table available on Google after Matchday 35, retrieved, april 18, 2026**.

```python
# Each team's season results stored as a dictionary
rivers_united = {
    "name": "Rivers United",
    "won": 16,
    "drawn": 10,
    "lost": 8,
    "goals_scored": 37,
    "goals_conceded": 27
}

enugu_rangers = {
    "name": "Enugu Rangers",
    "won": 15,
    "drawn": 11,
    "lost": 8,
    "goals_scored": 39,
    "goals_conceded": 24
}

ikorodu_city = {
    "name": "Ikorodu City",
    "won": 15,
    "drawn": 10,
    "lost": 9,
    "goals_scored": 37,
    "goals_conceded": 30
}

abia_warriors = {
    "name": "Abia Warriors",
    "won": 15,
    "drawn": 7,
    "lost": 12,
    "goals_scored": 28,
    "goals_conceded": 26
}

shooting_stars = {
    "name": "Shooting Stars",
    "won": 15,
    "drawn": 6,
    "lost": 13,
    "goals_scored": 28,
    "goals_conceded": 26
}
```

```python
#creating a list for the top teams
top_teams = [rivers_united,
             enugu_rangers,
             ikorodu_city,
             abia_warriors,
             shooting_stars]
```

 **The next series of code block will contain the  user-defined function that would perfectly take care of arithmetics for pointa accumulated so far, goal difference and total games played to complete the publicly available table, while win rate and average goal scored for further insight**

```python
# calculating the point accumulated so far
def calculate_points(team):
    return (team["won"] * 3) + (team["drawn"] * 1) # 3 points for the win, #1 points for the draw
```

```python
# Goal difference: goals scored minus goals conceded
def calculate_goal_difference(team):
    return team["goals_scored"] - team["goals_conceded"]
```

```python
# Total games played
def calculate_games_played(team):
    return team["won"] + team["drawn"] + team["lost"]
```

```python
# Win rate: wins divided by games played, multiplied by 100
def calculate_win_percentage(team):
    games = calculate_games_played(team)
    return (team["won"] / games) * 100
```

```python
# Average goals scored per game
def calculate_avg_goals(team):
    games = calculate_games_played(team)
    return round(team["goals_scored"] / games, 2) # rounded to 2 decimal places
```

#testing out the functions

```python
# Test on Ikorodu City
print(calculate_points(ikorodu_city))           # Should print 31
print(calculate_goal_difference(ikorodu_city))  # Should print 12
print(calculate_games_played(ikorodu_city))     # Should print 16
print(calculate_win_percentage(ikorodu_city))   # Should print 56.25
print(calculate_avg_goals(ikorodu_city))        # Should print 1.62
```

```python
# Build a results list — each item is a team's full stats
results = []

for team in top_teams:
    team_stats = {
        "name": team["name"],
        "played": calculate_games_played(team),
        "won": team["won"],
        "drawn": team["drawn"],
        "lost": team["lost"],
        "goals_for": team["goals_scored"],
        "goals_against": team["goals_conceded"],
        "goal_diff": calculate_goal_difference(team),
        "points": calculate_points(team),
        "win_pct": calculate_win_percentage(team),
        "avg_goals": calculate_avg_goals(team)
    }
    results.append(team_stats)
```

```python
def get_points(team):
    return team["points"]

results_sorted = sorted(results, key=get_points, reverse=True)
```

```python
# Print a proper league table header
print("=" * 75)
print(f"{'POS':<4} {'TEAM':<20} {'P':<4} {'W':<4} {'D':<4} {'L':<4} "
      f"{'GF':<4} {'GA':<4} {'GD':<5} {'PTS':<5} {'WIN%':<7} {'AVG'}")
print("=" * 75)

# Print each team's row
for position, team in enumerate(results_sorted, start=1):
    print(
        f"{position:<4} "
        f"{team['name']:<20} "
        f"{team['played']:<4} "
        f"{team['won']:<4} "
        f"{team['drawn']:<4} "
        f"{team['lost']:<4} "
        f"{team['goals_for']:<4} "
        f"{team['goals_against']:<4} "
        f"{team['goal_diff']:<+5} "     # the + shows + or - sign
        f"{team['points']:<5} "
        f"{team['win_pct']:<7.1f} "
        f"{team['avg_goals']}"
    )

print("=" * 75)
```

```python
print()
print("--- SEASON INSIGHTS ---")

# Find team with most goals scored
most_goals = results_sorted[0]["goals_for"]
best_attack_team = results_sorted[0]["name"]

for team in results_sorted:
    if team["goals_for"] > most_goals:
        most_goals = team["goals_for"]
        best_attack_team = team["name"]

print(f"Best attack:  {best_attack_team} with {most_goals} goals scored")

# Find team with fewest goals allowed
fewest_conceded = results_sorted[0]["goals_against"]
best_defence_team = results_sorted[0]["name"]

for team in results_sorted:
    if team["goals_against"] < fewest_conceded:
        fewest_conceded = team["goals_against"]
        best_defence_team = team["name"]

print(f"Best defence: {best_defence_team} with only {fewest_conceded} goals conceded")

# Show league leader (first team in sorted list)
leader = results_sorted[0]
print(f"League leader: {leader['name']} — {leader['points']} points")
```
