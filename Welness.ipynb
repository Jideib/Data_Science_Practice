# Wellness Score Logic and Limitations
## This is my day 4 project of building in public, learning from scratch - Data Science 30 days Holiday Sprint
### About the Wellness Score

The Wellness Score is a weighted average designed to provide a quick snapshot of a student's current lifestyle habits.

### The Formula:
`Score=(Sleep×2)+(Exercise×5)+(Water×3)−(Stress×2)`

### Limitations:

*   **Subjectivity:** Stress levels are self-reported and vary greatly between individuals.

*   **Nuance:** The tool does not account for the quality of sleep or the intensity of exercise.

*   **External Factors:** It ignores diet, social support, and pre-existing medical conditions.

```python
# List of 10 fictional student records in a dictionary
students = [
    {"name": "Yetunde", "sleep": 7, "exercise": 3, "stress": 4, "water": 6},
    {"name": "Tolu", "sleep": 5, "exercise": 1, "stress": 9, "water": 3},
    {"name": "Dare", "sleep": 8, "exercise": 5, "stress": 2, "water": 8},
    {"name": "Faith", "sleep": 4, "exercise": 0, "stress": 8, "water": 2},
    {"name": "Fidelix", "sleep": 6, "exercise": 2, "stress": 6, "water": 5},
    {"name": "Amina", "sleep": 9, "exercise": 4, "stress": 3, "water": 10},
    {"name": "George", "sleep": 5, "exercise": 2, "stress": 7, "water": 4},
    {"name": "Kawthar", "sleep": 7, "exercise": 3, "stress": 5, "water": 7},
    {"name": "Ibraheem", "sleep": 6, "exercise": 1, "stress": 8, "water": 3},
    {"name": "Olawunmi", "sleep": 8, "exercise": 6, "stress": 1, "water": 9},
]
```

```python
def calculate_wellness_score(student):
    """
    Reasoning for Formula:
    - Sleep (x2): Essential foundation, but usually has a capped benefit.
    - Exercise (x5): High weight because activity is a strong buffer against stress.
    - Water (x3): Good indicator of self-care and physical maintenance.
    - Stress (-x2): Stress is a negative weight that offsets healthy habits.
    """
    sleep_val = student["sleep"] * 2
    exercise_val = student["exercise"] * 5
    water_val = student["water"] * 3
    stress_penalty = student["stress"] * 2

    return sleep_val + exercise_val + water_val - stress_penalty
```

```python
def classify_student(score):
    if score < 25:
        return "Needs Support"
    elif 25 <= score <= 45:
        return "Doing OK"
    else:
        return "Thriving"
```

```python
def generate_report(student_list):
    print(f"{'Name':<12} | {'Score':<6} | {'Status':<15}")
    print("-" * 38)

    counts = {"Needs Support": 0, "Doing OK": 0, "Thriving": 0}

    for student in student_list:
        score = calculate_wellness_score(student)
        category = classify_student(score)

        # Update counts
        counts[category] += 1

        print(f"{student['name']:<12} | {score:<6} | {category:<15}")

    print("-" * 38)
    print("\nSummary Counts:")
    for cat, count in counts.items():
        print(f"{cat}: {count}")
```

```python
# Execute the report
generate_report(students)
```
